```python
df = movies.merge(credits, on='title')
```

```python
# Taking only the essential columns required for model creation

df = df[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
import ast

def extract_features(obj):
    return [row['name'] for row in ast.literal_eval(obj)]
```

```python
df['genres'] = df['genres'].apply(extract_features)
```

```python
df['genres'].head()
```

```python
df['keywords'] = df['keywords'].apply(extract_features)
```

```python
df.head()
```

```python
df['cast'][0]
```

```python
#  Extracting only the first 3 cast names, since most of the recommendation lie in these 3 names

def extract_first_3(obj):
    return [row['name'] for i,row in enumerate(ast.literal_eval(obj)) if i < 3]
```

```python
df['cast'] = df['cast'].apply(extract_first_3)
```

```python
df.head()
```

```python
ast.literal_eval(df['crew'][0])
```

```python
def extract_director(obj):
    return [row['name'] for row in ast.literal_eval(obj) if row['job'] == 'Director']
```

```python
df['crew'].head(1).apply(extract_director)
```

```python
df['crew'] = df['crew'].apply(extract_director)
```

```python
df.head()
```

```python
df['overview'][0]
```

```python
df['overview'] = df['overview'].apply(lambda x: x.split())
```

```python
df['genres'] = df['genres'].apply(lambda x:[i.replace(" ","") for i in x])
df['keywords'] = df['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
df['cast'] = df['cast'].apply(lambda x:[i.replace(" ","") for i in x])
df['crew'] = df['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
df.head()
```

```python
df['tags'] = df['overview'] + df['genres'] + df['keywords'] + df['cast'] + df['crew']
```

```python
df.head()
```

```python
new_df = df[['movie_id','title','tags']]
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df.head()
```

```python
import nltk

from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
def stem(text):
    stemmed_list = [ps.stem(word) for word in text.split()]
    
    return " ".join(stemmed_list)
```

```python
new_df['tags'].apply(stem)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
cv.fit_transform(new_df['tags']).toarray().shape
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
len(cv.get_feature_names())
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
new_df[new_df['title'] == 'Avatar'].index[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x: x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x: x[1])[1:6]
    
    for movie in movies_list:
        print(new_df.iloc[movie[0]].title)
```

```python
recommend('Avatar')
```
